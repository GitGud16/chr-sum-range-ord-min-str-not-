```python
from netCDF4 import Dataset

# فتح مجموعة البيانات باستخدام netCDF4
opendap_url            = 'C:\Users\3ab3zeez\Downloads\AERDB_L2_VIIRS_NOAA20.A2024260.0948.002.2024260220657 (1).n'
#opendap_url = 'https://airsl2.gesdisc.eosdis.nasa.gov/opendap/Aqua_AIRS_Level2/AIRX2RET.006/2016/249/AIRS.2016.09.05.001.L2.RetStd.v6.0.31.0.G16249172736.hdf'

# فتح الملف ومحاولة الوصول إلى المتغيرات
try:
    dataset = Dataset(opendap_url)
    print("تم الوصول إلى مجموعة البيانات بنجاح!")

    # عرض قائمة المتغيرات المتاحة
    available_vars = list(dataset.variables.keys())
    print("المتغيرات المتاحة في مجموعة البيانات:\n", available_vars)

    # محاولة الوصول إلى المتغيرات بتحميل جزء صغير جدًا
    # تحديد المتغيرات التي ترغب في الوصول إليها، يمكنك تغيير 'TSurfStd' أو 'TAirStd' إلى أي متغير آخر من القائمة
    if 'TSurfStd' in available_vars:
        # عرض أبعاد المتغير
        print(f"\nأبعاد متغير TSurfStd: {dataset.variables['TSurfStd'].shape}")
        
        # تحميل جزء صغير جدًا (أول 5 قيم) للتحقق من إمكانية الوصول
        small_subset = dataset.variables['TSurfStd'][0:5, 0:5]
        print(f"أول 5 قيم من TSurfStd:\n{small_subset}")
    else:
        print("المتغير 'TSurfStd' غير متاح في مجموعة البيانات.")
        
    # التحقق من متغير آخر مثل 'TAirStd' (درجة حرارة الهواء)
    if 'TAirStd' in available_vars:
        print(f"\nأبعاد متغير TAirStd: {dataset.variables['TAirStd'].shape}")
        small_subset = dataset.variables['TAirStd'][0:5, 0:5]
        print(f"أول 5 قيم من TAirStd:\n{small_subset}")
    else:
        print("المتغير 'TAirStd' غير متاح في مجموعة البيانات.")

except RuntimeError as e:
    print(f"حدث خطأ أثناء محاولة الوصول إلى البيانات: {e}")
```
